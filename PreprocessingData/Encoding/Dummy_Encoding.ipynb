```python
 import pandas as pd
```

```python
data = {
    'Employee_id' : [1,2,3,4,5,6],
    'Gender' : ['M', 'F', 'M', 'M','M','F'],
    'Mode' : ['Good', 'Nice', 'Good', 'Great', 'Nice', 'Great']
}
df =  pd.DataFrame(data)
```

```python
column_name  = df.select_dtypes(include = 'object').columns.to_list()
column_name
```

```python
encoded_data = pd.get_dummies(df[column_name], drop_first= True , dtype=float)
ecoded_data = pd.concat([df, encoded_data], axis = 1)
ecoded_data
```

```python
ecoded_data = ecoded_data.drop(column_name, axis = 1)
ecoded_data
```

*  One-Hot Encoding: More explicit, representing each category with its own column. Can be useful for models that require explicit feature representation or for scenarios where category ordering doesn’t matter.

* Dummy Encoding: More compact, reducing the number of new columns. Preferred for linear regression models to avoid multicollinearity issues.

**The best choice often depends on your specific use case and the model you’re using. However, there’s a general rule of thumb:**

* Linear Regression: Use dummy encoding to avoid the dummy variable trap.
* Other Algorithms (Decision Trees, Random Forests, etc.): Either one-hot or dummy encoding can work well.
